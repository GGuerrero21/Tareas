Use $10$ puntos equidistantes para encontrar un polinomio interpolante de la función $f(x) = \tanh(x)$ entre $x = -10$ y $x = 10$. Compare el polinomio con la función obtenida. Explique.

-------------------------------------------------------------------------------------------------------------------

```python
#Como siempre, primero importamos las librerias a utilizar
import numpy as np
import matplotlib.pyplot as plt

#Para interpolar usare la funcion de Lagrange presentada en clases
def lagrange(x, puntos):
    '''
    Interpola una función que pasa por n puntos como un polinomio
    de Legendre y lo evalúa en el punto x. Los puntos deben ser
    dados como una lista de tuplas (xi, f(xi)).
    '''

    Lis = np.ones(len(puntos))
    fs = np.ones(len(puntos))
    for i in range(len(puntos)):
        fs[i] = puntos[i][1]
        for j in range(len(puntos)):
            if j != i:
                Lis[i] *= (x - puntos[j][0])/(puntos[i][0] - puntos[j][0])

    return sum(fs*Lis)


intervalo = np.linspace(-10,10,10) #Intervalo de puntos a utilizar
puntos = np.array([(i,np.tanh(i)) for i in intervalo]) #Valor de tanh(x) para cada punto en el intervalo anterior

x = np.linspace(-10, 10, 100) #Eje x para graficar
aprox = [lagrange(xi,puntos) for xi in x] #Interpolacion de tanh(x)
exacta = np.tanh(x) #Valor exacto(en Python) para la funcion solicitada

#Graficamos el valor exacto de tanh(x) y la interpolacion de Lagrange con los puntos solicitados
plt.scatter(puntos[:,0],puntos[:,1],color='black') #Grafica de los puntos
plt.plot(x, aprox, ls='--', label='Interpolada n=10') #Grafica de la interpolacion
plt.plot(x, exacta, label='Exacta') #Grafica de la funcion exacta
#Etiquetas para los eje
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Interpolacion tanh(x)')
plt.grid(True)
plt.legend()
plt.show() 
```

Como se puede observar, tal como fue visto en clases, hay un intervalo (pequeño) en el que la interpolacion tiene un valor muy similar al de la funcion exacta, sin embargo, fuera de ese intervalo la curva de los valor adquiere un error considerable comparandola con la funcion exacta.
