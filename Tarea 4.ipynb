Un colega le pasa a usted un archivo .hd5 que contiene tres "datasets" 'x', 'y', 'e'. Todos los datasets tienen la misma longitud y el i-ésimo elemento de cada dataset corresponde a una propiedad diferente del i-ésimo dato.

Escriba una función que grafique esos resultados donde 'x' es el eje horizontal, 'y' el eje vertical y 'e' son las barras de error de 'y'.

CUIDADO: Recuerde que el dataset es un puntero y deja de existir cuando se cierra el archivo.

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
#Importamos los paquetes a utilizar
import matplotlib.pyplot as plt #Matplotlib para hacer el grafico solicitado 
import numpy as np #Numpy para generar los numeros
import h5py #H5py para crear el archivo recibido
```

```python
np.random.seed(21) #Definimos la semilla para que los valores no cambien cada vez que el codigo corra

#Creo un archivo simulando el enviado por el colega
with h5py.File("datasets.hd5", "w") as file: #Lo definimos con el nombre de datasets y lo abrimos en modo escritura
    x = file.create_dataset("x", data=np.arange(20)) #Creamos los datos del eje "x" del 0 al 19
    y = file.create_dataset("y", data=np.random.randint(20, size=20)) #Para el eje "y" creamos 20 datos aleatoriamente
    e = file.create_dataset("e", data=np.random.randint(20, size=20)) #Para los datos de "e" que serian las barras de error creamos nuevamente 20 datos aleatorios

#Con el archivo ya creado, lo abrimos en modo lectura para utilizar los datos
with h5py.File("datasets.hd5", "r") as file:
    x = file["x"][...] #Definimos una variable que contenga el array con los valores de "x"
    y = file["y"][...] #Lo mismo para los diferentes valores de "y"
    e = file["e"][...] #Por ultimo para los datos de "e"

#Creamos el gráfico
plt.errorbar(x,y,e, #Con formato de barra de error que fue el solicitado
             ecolor="black", #Color de la barra de error negro
             elinewidth = 0.5, #Anchura de la barra de error
             linewidth = 1.5, #Anchura de la linea representativa de los datos
             capsize= 2.0) #Tamaño de las lineas de tobe de las barras de error

#Como no tenemos unidades o que signifacria cada eje, lo dejare con un nombre generico
plt.xlabel("x") #Ponemos nombre al eje x
plt.ylabel("y") #Ponemos nombre al eje y

#Solicitamos que nos muestre el gráfico
plt.show()
```
